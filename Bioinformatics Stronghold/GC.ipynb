```python
with open('rosalind_gc.txt', 'rt') as file:
    dic = {}
    FASTA_id = None
    for line in file:
        line = line.strip()
        if line.startswith(">"):
            FASTA_id = line
            dic[FASTA_id[1:]] = ""
        else:
            dic[FASTA_id[1:]] += line

print(dic)        
```

```python
highest_gc = 0
for key in dic.keys():
    gc_count = 0
    for letter in dic[key]:
        if letter == "G" or letter == "C":
            gc_count += 1
    gc_content = gc_count/len(dic[key])
    if highest_gc < gc_content:
            saved_id = key
            highest_gc = gc_content
print(f"{saved_id}\n{highest_gc*100}")
```
